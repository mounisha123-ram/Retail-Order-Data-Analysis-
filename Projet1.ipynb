# **3) Trimming Spaces: Remove trailing spaces from text fields.**

Explanation for the code:-

Breakdown:
applymap()

This function applies a transformation element-wise to every value in the DataFrame.
It works on each individual cell, unlike apply() which works on columns or rows.
lambda x: x.strip() if isinstance(x, str) else x

This is an anonymous function (lambda function) that:
Checks if the value (x) is a string (isinstance(x, str)).
If x is a string, it removes leading and trailing spaces using .strip().
If x is not a string (e.g., a number), it remains unchanged.

```python
# Remove leading & trailing spaces from all string columns
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

# **4) #CALCULATING DISCOUNTS---derive new columns discount , sale price and profit**

Calculate Discount, Sale Price & Profit
👉 Deriving New Columns:
Discount Amount = (list_price * discount_percent) / 100
Sale Price = list_price - discount_amount
Profit = (sale_price - cost_price) * quantity

```python
#.astype(int) ensures all values are stored as integers instead of floats.
#If there are NaN (missing) values, use .fillna(0).astype(int) to replace NaN with 0 before conversion.

df["discount_amount"] = ((df["list_price"] * df["discount_percent"]) / 100).astype(int)

# Calculate Sale Price
df["sale_price"] = (df["list_price"] - df["discount_amount"]).astype(int)

# Calculate Profit
df["profit"] = ((df["sale_price"] - df["cost_price"]) * df["quantity"]).astype(int)

# Display first few rows with new columns
df[["list_price", "discount_percent", "discount_amount", "sale_price", "cost_price", "quantity", "profit"]].head()
#df.head()
```

# **Data Analysis with SQL**
# **Business Insights through SQL Queries:**


Top-Selling Products: Identify the products that generate the highest revenue based on sale prices.
Monthly Sales Analysis: Compare year-over-year sales to identify growth or decline in certain months.
Product Performance: Use functions like GROUP BY, HAVING, ROW_NUMBER(), and CASE WHEN to categorize and rank products by their revenue, profit margin, etc.
Regional Sales Analysis: Query sales data by region to identify which areas are performing best.
Discount Analysis: Identify products with discounts greater than 20% and calculate the impact of discounts on sales.


```python
#Top-Selling Products: Identify the products that generate the highest revenue based on sale prices.
mycursor.execute(""" SELECT product_id,
       SUM(sale_price * quantity) AS total_revenue
FROM project.orders
GROUP BY product_id
ORDER BY total_revenue DESC
LIMIT 5
     """)

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
# Monthly Sales Analysis: Compare year-over-year sales to identify growth or decline in certain months.
mycursor.execute(""" SELECT YEAR(order_date) AS year,
       MONTH(order_date) AS month,
       SUM(sale_price * quantity) AS total_sales
FROM project.orders
GROUP BY YEAR(order_date), MONTH(order_date)
ORDER BY year, month
     """)

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#Product Performance: Use functions like GROUP BY, HAVING, ROW_NUMBER(), and CASE WHEN to categorize and rank products by their revenue, profit margin, etc.
mycursor.execute(""" SELECT product_id,
       SUM(sale_price * quantity) AS total_revenue,
       SUM(profit) AS total_profit,
       RANK() OVER (ORDER BY SUM(sale_price * quantity) DESC) AS revenue_rank,
       RANK() OVER (ORDER BY SUM(profit) DESC) AS profit_rank
FROM project.orders
GROUP BY product_id limit 10
     """)

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#Regional Sales Analysis: Query sales data by region to identify which areas are performing best.
mycursor.execute(""" SELECT region,
       SUM(sale_price * quantity) AS total_sales
FROM project.orders
GROUP BY region
ORDER BY total_sales DESC
     """)

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#Discount Analysis: Identify products with discounts greater than 20% and calculate the impact of discounts on sales.
mycursor.execute("""SELECT product_id,
       discount_percent,
       SUM(sale_price * quantity) AS total_sales,
       CASE
           WHEN discount_percent > 20 THEN 'High Discount'
           WHEN discount_percent BETWEEN 10 AND 20 THEN 'Medium Discount'
           ELSE 'Low Discount'
       END AS discount_category
FROM project.orders
GROUP BY product_id, discount_percent
ORDER BY discount_percent DESC limit 10
     """)

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#1. Get all order details with product info
mycursor.execute("""  SELECT o.order_id, o.order_date, o.ship_mode,
       p.product_id, p.category, p.sub_category,
       op.quantity, op.sale_price, op.profit
FROM project.order_product op
JOIN project.order o ON op.order_id = o.order_id
JOIN project.product p ON op.product_id = p.product_id limit 10
     """)

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
 #2.Total sales for each category
mycursor.execute(""" SELECT p.category, SUM(op.sale_price) AS total_sales
FROM project.order_product op
JOIN project.product p ON op.product_id = p.product_id
GROUP BY p.category
 """)

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#3. Top 5 products with highest profit
mycursor.execute(""" SELECT p.product_id, p.category,p.sub_category, SUM(op.profit) AS total_profit
FROM project.order_product op
JOIN project.product p ON op.product_id = p.product_id
GROUP BY p.product_id, p.sub_category
ORDER BY total_profit DESC
LIMIT 5;
 """)

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#4. Number of orders placed in each region
mycursor.execute(""" SELECT o.region, COUNT(DISTINCT o.order_id) AS total_orders
FROM project.orders o
JOIN project.order_product op ON o.order_id = op.order_id
GROUP BY o.region;

 """)

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#5.  Find the top 5 cities with the highest total sales.
mycursor.execute(""" SELECT o.city,
       SUM(op.sale_price) AS total_sales
FROM project.order_product op
JOIN project.order o ON op.order_id = o.order_id
GROUP BY o.city
ORDER BY total_sales DESC
LIMIT 5;

 """)

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
# 6.Total discount given for each category
mycursor.execute(""" SELECT p.category, SUM(op.discount_amount) AS total_discount
FROM project.order_product op
JOIN project.product p ON op.product_id = p.product_id
GROUP BY p.category;

 """)

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#7. Orders with the highest discount percentage
mycursor.execute(""" SELECT o.order_id,p.category,op.product_id, op.discount_percent
FROM project.order_product op
JOIN project.orders o ON op.order_id = o.order_id
JOIN project.product p ON op.product_id = p.product_id
ORDER BY op.discount_percent DESC
LIMIT 10;

 """)

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#8. Average sale price for each sub-category
mycursor.execute(""" SELECT p.sub_category, AVG(op.sale_price) AS avg_sale_price
FROM project.order_product op
JOIN project.product p ON op.product_id = p.product_id
GROUP BY p.sub_category;

 """)

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#9. Most popular product (ordered most times)
mycursor.execute(""" SELECT p.product_id, p.sub_category, COUNT(op.order_id) AS order_count
FROM project.order_product op
JOIN project.product p ON op.product_id = p.product_id
GROUP BY p.product_id, p.sub_category
ORDER BY order_count desc
limit 10;

 """)

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#10. Total revenue, profit, and discount given for each region
mycursor.execute(""" SELECT o.region, SUM(op.sale_price) AS total_revenue,
       SUM(op.profit) AS total_profit, SUM(op.discount_amount) AS total_discount
FROM project.orders o
JOIN project.order_product op ON o.order_id = op.order_id
GROUP BY o.region;

 """)

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```
